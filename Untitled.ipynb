***Part 3.1***

```python
from libsvm.svmutil import *
import csv
import numpy as np
```

***Part 3.2***

```python
with open('abalone.data', 'r') as file:
    data=list(csv.reader(file))
    data=np.array(data)

np.random.shuffle(data[0:3133,:])
y=data[:,-1].astype(np.int)
X=data[:,1:-1].astype("float")
gender=data[:,0]
gender=np.array([gender])
gender[gender=='F']=1
gender[gender=='I']=0
gender[gender=='M']=2
gender=gender.astype(np.int)
gender[gender==2]=-1

rand_data=np.concatenate((gender.T, X), axis=1)

y[y<=9]=1
y[y>9]=0
y=y.astype(np.int)
#y[y==0]=-1
```

```python
train_data=rand_data[:3133]
train_label=y[:3133]
test_data=rand_data[3133:]
test_label=y[3133:]

mins=np.min(train_data,0)
maxs=np.max(train_data,0)
train_data=(train_data[:]-mins)/(maxs-mins)
test_data = (test_data[:]-mins)/(maxs-mins)
```

***Part 3.3***

```python
import matplotlib.pyplot as plt

group_elem=int(len(train_data)/5)

d_values=[1,2,3,4,5]
kernel_type=1
k=6
for d_value in d_values:
    means=np.zeros(2*k+1)
    mean_p_std=np.zeros(2*k+1)
    mean_m_std=np.zeros(2*k+1)
    for m in range(-k,k+1):
        C=3 ** m
        sum_S_value=0
        acc=0
        error=[]
        for i in range(5):
            cv_test_d=train_data[i*group_elem:(i+1)*group_elem]
            cv_test_l=train_label[i*group_elem:(i+1)*group_elem]
            cv_train_d=np.delete(train_data, slice(i*group_elem,(i+1)*group_elem,1), 0)
            cv_train_l=np.delete(train_label, slice(i*group_elem,(i+1)*group_elem,1), 0)   
        
        
            model = svm_train(cv_train_l, cv_train_d, '-c '+ str(C) + ' -t ' + str(kernel_type) + ' -d '+ str(d_value) + ' -q')
            S_value=len(model.get_SV())
            p_label, p_acc, p_val = svm_predict(cv_test_l, cv_test_d, model , '-q')
            
            sum_S_value=sum_S_value+S_value
            a=np.sum((p_label-cv_test_l)**2)/len(p_label)
            error.append(1-p_acc[0]/100)
            acc=acc+1-p_acc[0]/100

            #print('SVM kernel=' + str(kernel_type) + ' C=' + str(m) + ' acc=' + str(1-p_acc[0]/100) + ' n=' +str(S_value))
        #print(' acc=' + str(acc/5) + ' n=' +str(np.floor(sum_S_value/5)))
        means[m+k]=np.mean(error)
        mean_p_std[m+k]=means[m+k]+np.std(error)
        mean_m_std[m+k]=means[m+k]-np.std(error)
    #print()
    #print(d_value)
    #print(means)
    #print(mean_p_std)
    #print(mean_m_std)
    #print()
    
    plt.figure()
    plt.plot(range(-k,k+1),means)
    plt.plot(range(-k,k+1),mean_p_std)
    plt.plot(range(-k,k+1),mean_m_std)
    plt.xlabel('Log_3 C')
    plt.ylabel('cross validation error')
    plt.title('cross validation +- std for d=' + str(d_value))
```
